```python
import numpy as np
import pandas as pd
import matplotlib as plt
from statsmodels.tsa.arima_model import ARIMA
%matplotlib inline
```

```python
df_original = pd.read_csv('df/Index2018.csv')
df = df_original.copy()
df['date'] = pd.to_datetime(df['date'], dayfirst = True)
df.set_index('date', inplace=True)
df = df.asfreq('b')
df = df.fillna(method='ffill')
```

```python
dfexog = df[['spx', 'dax', 'nikkei']]

model = ARIMA(df.ftse, exog = dfexog, order=(1,1,1))
results = model.fit()
results.summary()
```
